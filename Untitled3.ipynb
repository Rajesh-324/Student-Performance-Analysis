# Student Performance Analysis

This project analyzes student performance data using Python.
The dataset is cleaned, analyzed, and visualized using pandas and matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Dataset

In this step, the student performance dataset is loaded using pandas.


```python
df = pd.read_csv("student-por[1].csv", sep=';')
df.head()
```

## Exploring Dataset

Here we check dataset shape, data types, missing values, and duplicates.

```python
print(df.shape)
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

## Removing Duplicates

Duplicate rows are removed from the dataset.

```python
df = df.drop_duplicates()
print(df.shape)
```

## Average Final Grade

Calculating the average final grade (G3) of students.

```python
print(df['G3'].mean())
```

## Students Scored Above 15

Counting students who scored more than 15 marks.

```python
print((df['G3'] > 15).sum())
```

## Correlation Between Study Time and Performance

Checking whether study time affects student performance.

```python
print(df['studytime'].corr(df['G3']))
```

## Gender-wise Average Performance

Comparing average scores of male and female students.

```python
print(df.groupby('sex')['G3'].mean())
```

## Histogram of Final Grades

This graph shows the distribution of student final grades.

```python
plt.hist(df['G3'])
plt.xlabel("Final Grade")
plt.ylabel("Count")
plt.title("Distribution of Final Grades")
plt.show()
```

## Scatter Plot: Study Time vs Final Grade

This graph shows the relationship between study time and final grades.

```python
plt.scatter(df['studytime'], df['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.title("Study Time vs Final Grade")
plt.show()
```

## Bar Chart: Male vs Female Average Score

This chart compares average final scores of male and female students.

```python
df.groupby('sex')['G3'].mean().plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.title("Male vs Female Average Score")
plt.show()
```

## Conclusion

The dataset was successfully cleaned and analyzed using Python.

Students with higher study time generally performed better.
The visualizations helped understand grade distribution and gender-wise performance.
